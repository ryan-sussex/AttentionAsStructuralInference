```python
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = "browser"

import json
import numpy as np
import pandas as pd
from plotly.subplots import make_subplots
```

```python
with open("expanding.json", mode="r") as f:
    training_history = json.load(f)
```

```python
training_history
```

```python
dims = 10
n = 100
iterations = np.array(training_history["expanding"]["iters"]) + 1
window = np.array(training_history["expanding"]["window"])

expanding_complexity = iterations  +  window * dims
print(expanding_complexity)

expanding_complexity = np.mean(expanding_complexity)
expanding_complexity
```

```python
complexity = n * dims
complexity
```

```python
traces = []
start = 0
x = list(range(start, len(training_history["expanding"]["loss"])))
# for i in training_history:
expanding_loss_trace = go.Scatter(
            x=x,
            y=training_history["expanding"]["loss"],
            name=f"SELF-ATTENTION",        
            #     error_y=dict(
            #     type='data', # value of error bar given in data coordinates
            #     array=standard_std * 0.5,
            #     visible=True
            # )
        )
standard_loss_trace = go.Scatter(
            x=x,
            y=training_history["standard"]["loss"],
            name=f"STANDQARD",
            # error_y=dict(
            #     type='data', # value of error bar given in data coordinates
            #     array=long_std * 0.5,
            #     visible=True
            # )
        )
layout = go.Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)',
    xaxis_title="Batch",
    yaxis_title="Normalised MSE",
    legend=dict(x=0.6, y=1)
    # legend_x=0, 
    # legend_y=0,
)


# )
```

```python
complexity_trace = go.Bar(x=["standard", "expanding"], y=[complexity, expanding_complexity])
```

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(expanding_loss_trace)
fig.add_trace(standard_loss_trace)
fig.add_trace(complexity_trace,secondary_y=True)
fig.show()
# fig.add_trace
```
