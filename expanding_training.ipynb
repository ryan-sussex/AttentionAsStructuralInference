```python
import plotly.graph_objects as go
import plotly.io as pio
pio.renderers.default = "browser"

import json
import numpy as np
import pandas as pd
from plotly.subplots import make_subplots
```

```python
with open("expanding.json", mode="r") as f:
    training_history = json.load(f)
```

```python
purple = "rgba(204,204,245,255)"
green = "rgba(191,230,191,255)"
```

```python
training_history
```

```python
print(np.mean(training_history["expanding"]['iters'][-50:]))
print(np.mean(training_history["expanding"]['iters'][:50]))
```

```python
dims = 10
n = 50
iterations = np.array(training_history["expanding"]["iters"]) + 1
window = np.array(training_history["expanding"]["window"])

expanding_complexity_series =   np.maximum(28, np.minimum(window,n)) * (dims + iterations)
# print(expanding_complexity)

expanding_complexity = np.mean(expanding_complexity_series)
expanding_complexity
```

```python
expanding_loss = np.array(training_history["expanding"]['loss']).reshape(10, -1).mean(axis=1)
standard_loss = np.array(training_history["standard"]['loss']).reshape(10, -1).mean(axis=1)
```

```python
complexity = n * dims
complexity
```

```python
print(expanding_complexity/complexity)
```
